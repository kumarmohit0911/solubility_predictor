```python
compound_list = data['smiles']
```

```python
from rdkit.Chem import Descriptors
```

```python
# here i will be generating features with the help of rdkit descriptors
all_features = []
for compound in compound_list:
    mol = Chem.MolFromSmiles(compound)
    if not  mol:
        raise ValueError("error not a valid smile")
    else:
        features  = [
                    #Descriptors.MinVertexDegree(mol),
                    Descriptors.MolWt(mol),
                    Descriptors.NumHDonors(mol),
                    Descriptors.RingCount(mol),
                    Descriptors.NumRotatableBonds(mol),
                    Descriptors.TPSA(mol)
                ]
    all_features.append(features)
all_features
```

```python
col = ['Molecular weights','Number of H-Bond Donors','Number of Rings','Number of Rotable Bonds','Topical Polar Surface Area']
DF=pd.DataFrame(all_features,columns=col)
DF
```

```python
DF['measured log solubility in mols per litre'] = dtfm
```

```python
print (DF.shape)
DF.head()
```

```python
print(DF.shape)
for col in DF.columns:
    Q1 = DF[col].quantile(0.25)
    Q3 = DF[col].quantile(0.75)
    IQR = Q3 - Q1
    UL = Q3 + 1.5*IQR
    LL = Q1 - 1.5*IQR

    DF = DF[(DF[col] >= LL) & (DF[col] <= UL)]

    print(DF.shape)
        
```

```python
X=DF.iloc[:,[0,1,2,3,4]]
X.shape
```

```python
y=DF.iloc[:,[5]]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state = 42)
```
